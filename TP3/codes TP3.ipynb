```python
# Définition d'une classe GestionCptB pour gérer un compte bancaire
class GestionCptB:
    def __init__(self, nCompte, nomClient, soldeCompte):
        # Constructeur de la classe, initialise les attributs
        self.nCompte = nCompte  # Numéro de compte
        self.nomClient = nomClient  # Nom du client
        self.soldeCompte = soldeCompte  # Solde du compte

    def Versement(self, montant):
        # Méthode pour effectuer un versement
        self.soldeCompte += montant  # Augmente le solde du compte de montant

    def Retrait(self, montant):
        # Méthode pour effectuer un retrait
        if montant <= self.soldeCompte:
            self.soldeCompte -= montant  # Réduit le solde du compte de montant si le solde est suffisant
        else:
            print("Solde insuffisant.")  # Affiche un message si le solde est insuffisant

    def affichageinfo(self):
        # Méthode pour afficher les informations du compte
        print("Numéro de compte :", self.nCompte)
        print("Nom du client :", self.nomClient)
        print("Solde du compte :", self.soldeCompte)

# Créer un compte
compte1 = GestionCptB("123456", "John Doe", 1000.0)

# Effectuer un versement
compte1.Versement(5000)

# Effectuer un retrait
compte1.Retrait(2300)

# Afficher les informations du compte
compte1.affichageinfo()
```

```python
# Importer la bibliothèque numpy pour utiliser les constantes mathématiques
import numpy as np

# Définition de la classe Cercle
class Cercle:
    def __init__(self, rayon):
        # Constructeur de la classe, initialise l'attribut rayon
        self.rayon = rayon

    def aire(self):
        #methode pour calculer l'aire d'un cercle 
        A = np.pi * self.rayon**2
        return A
    
    def perimetre(self):
        #methode pour calculer le perimetre d'un cercle 
        P = 2 * np.pi * self.rayon 
        return P
    

# Créer un cercle avec un rayon de 5
circle = Cercle(10)

# Calculer l'aire du cercle
A = circle.aire()
print("Aire du cercle :", A)

# Calculer le périmètre du cercle
P = circle.perimetre()
print("Périmètre du cercle :", P)
```
